Voici le phénomène de Runge pour l'interpolation lagrangienne.
 
Même pour une fonction très régulière ($C^{\infty}(\mathbb R)$ !) les choses ne se passent pas comme on pourrait s'imaginer...

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

def lag(n):
    x = np.linspace(-1, 1, n)
    y = 1/(1+x**2)
    return lagrange(x, y)

if __name__ == "__main__":
    x_plt = np.linspace(-1.5, 1.5, 50)
    for n in range(3,8):
        poly = lag(n)
        plt.plot(x_plt, poly(x_plt), label=f"{n=}", linestyle='-')
    plt.plot(x_plt, 1/(1+x_plt**2), label=r'$\frac{1}{1+x^2}$', color='k')
    plt.legend()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

legs = []
for i in range(5):
    legs.append(np.polynomial.legendre.Legendre.basis(deg=i, domain=[-1, 1]))

def plot_poly(pols, title="", dom=[-1, 1]):
    t = np.linspace(dom[0], dom[1], 200)
    for k,pol in enumerate(pols):
        plt.plot(t, pol(t), label=f"{k=}")
    plt.legend()
    plt.title(title)
    plt.show()
plot_poly(legs, title="Legendre Polynomials")
```

```python
chebs = []
for i in range(5):
    chebs.append(np.polynomial.chebyshev.Chebyshev.basis(deg=i, domain=[-1, 1]))
plot_poly(chebs, title="Chebychev Polynomials")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial.laguerre import Laguerre

def plot_laguerre(n_max=5, xmax=10):
    x = np.linspace(0, xmax, 500)
    
    for n in range(n_max):
        L = Laguerre.basis(n)   # L_n(x)
        plt.plot(x, L(x), label=f"L_{n}")
    
    plt.title("Laguerre polynomials")
    plt.legend()
    plt.show()

plot_laguerre(n_max=6, xmax=8)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial.hermite import Hermite

def plot_hermite(n_max=5, xmax=4):
    x = np.linspace(-xmax, xmax, 500)
    
    for n in range(n_max):
        H = Hermite.basis(n)   # H_n(x)
        plt.plot(x, H(x), label=f"H_{n}")
    
    plt.title("Hermite polynomials")
    plt.legend()
    plt.show()

plot_hermite(n_max=7, xmax=1.5)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import eval_jacobi, roots_jacobi

def plot_jacobi(n_max=5, alpha=0.0, beta=0.0):
    x = np.linspace(-1, 1, 500)
    for n in range(n_max):
        y = eval_jacobi(n, alpha, beta, x)
        y1 = eval_jacobi(n, alpha, beta, 1.0)
        plt.plot(x, y/y1, label=f"P_{n}")
    plt.title(f"Jacobi polynomials α={alpha}, β={beta}")
    plt.legend()
    plt.show()

plot_jacobi(n_max=6, alpha=1, beta=0.)     
plot_jacobi(n_max=6, alpha=0, beta=-0.5)    # Chebyshev, up to scaling
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial.legendre import Legendre
from numpy.polynomial.chebyshev import Chebyshev

n = 8

# degree-n polynomials
P = Legendre.basis(n, domain=[-1, 1])
T = Chebyshev.basis(n, domain=[-1, 1])

# zeros
x_leg = P.roots()
x_cheb = T.roots()

print("Legendre zeros:")
print(x_leg)
print()
print("Chebyshev zeros:")
print(x_cheb)

# plot the polynomials and their zeros
t = np.linspace(-1, 1, 1000)

plt.plot(t, P(t), label=f"Legendre P_{n}", linewidth=1, linestyle="-")
plt.plot(t, T(t), "--", label=f"Chebyshev T_{n}", linewidth=1, linestyle="-")

plt.plot(x_leg, np.zeros_like(x_leg), "o", label="Legendre zeros")
plt.plot(x_cheb, np.zeros_like(x_cheb), "x", label="Chebyshev zeros")

plt.axhline(0, linewidth=0.8)
plt.legend()
plt.title(f"Zeros of degree {n} polynomials")
plt.show()
```
